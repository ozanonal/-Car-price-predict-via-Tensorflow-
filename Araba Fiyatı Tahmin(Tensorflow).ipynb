Burada gÃ¶rÃ¼ldÃ¼ÄŸÃ¼ Ã¼zere %30' luk bir kÄ±smÄ± teste ayÄ±rmÄ±ÅŸ bulunmaktayÄ±m.

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense
```

```python
x_train.shape
```

Åžu anda burada 7393 verimiz ve 5 tane Ã¶zelliÄŸimiz var. Bu durumda 5 tane layerden baÅŸlatmamÄ±z uygun olur diye dÃ¼ÅŸÃ¼nerek iÅŸleme devam edeceÄŸim.

4 tane derin aÄŸ olacak ÅŸekilde katman ekleyeceÄŸim.

```python
model = Sequential()

model.add(Dense(5, activation = "relu"))
model.add(Dense(5, activation = "relu"))
model.add(Dense(5, activation = "relu"))
model.add(Dense(5, activation = "relu"))

model.add(Dense(1))

model.compile(optimizer = "adam", loss="mse")
```

```python
model.fit(x=x_train, y=y_train, validation_data = (x_test,y_test), batch_size = 250, epochs=310)
```

```python
lost = pd.DataFrame(model.history.history)
```

```python
lost.head()
```

```python
lost.plot()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
predict1 = model.predict(x_test) 
```

```python
predict1
```

```python
mean_absolute_error(y_test, predict1)
```
